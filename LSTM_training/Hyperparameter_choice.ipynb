```python
from glob import glob
import numpy as np
import pandas as pd
```

```python
all_params = glob("hyperparameters/*/")
```

```python
df = pd.DataFrame(columns=['parameters', 'max_FP', 'median_TP', 'median_FP', 'all_TP', 'all_FP'])
for parameter_set in all_params:
    try:
        with open(parameter_set+'test.txt') as f:
            for i in range(10):
                f.readline()
            tp = f.readline()
            fp = f.readline()
        false_positives = np.fromstring(fp[3:-1], sep=' ', count=6)
        true_positives = np.fromstring(tp[3:-1], sep=' ', count=6)
        df = df.append({
            'parameters': parameter_set,
            'max_FP': max(false_positives[1:]),
            'median_TP': np.median(true_positives[1:]),
            'median_FP': np.median(false_positives[1:]),
            'all_TP': true_positives[1:],
            'all_FP': false_positives[1:]
        }, ignore_index=True)
    except FileNotFoundError:
        continue
```

```python
df[df['max_FP']<0.05].sort_values(by=['median_TP', 'max_FP'], ascending=[False, True])
```
